```python
import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread("/content/circles.png")

if image is None:
    print("Error: Image not found.")
else:
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    resized = cv2.resize(gray, (300, 300))
    cv2.imwrite("processed.jpg", resized)

    print("Image processed and saved as processed.jpg")

    cv2_imshow(resized)
```
